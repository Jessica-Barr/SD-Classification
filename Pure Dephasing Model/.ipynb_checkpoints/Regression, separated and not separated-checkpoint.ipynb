```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
X = np.loadtxt('Data/Xtrainx_separated.csv', delimiter=',')
Y = np.loadtxt('Data/Ytrain_separated.csv', delimiter=',')
#YtrainPD is an array of shape n_training x 6. if the ith training example is sub_ohmic then element [i,0] = 1, if 
#it is Ohmic then [i,1] = 1, and if it is super-Ohmic then [i,2] = 1. #element [i,3] is set equal to 
#$\omega_c$ while elements [i,4] and [i,5] are set equal to s and $\eta$, respectively.
```

```python
#function to take X and Y and return a training, test and validation set containing the Fourier coefficients, along
#with the corresponding labels. The last 3 columns of Y contain the values of $'\eta'$, $\omega_c$ and $s$ so we 
#only use the first three columns. Likewise, at t=0 all of the signals are equal, so we omit the first timestep.
#Ntrain is the number of trajectories in the training set, Nval is the number in the validation set and Ntest is 
#the number in the test set
def fouriertrainvaltest(X, Y, Ntrain, Nval, Ntest):
    
    #Generating a training set with Ntrain trajectories, a validation with Nval trajectories and a test set with
    #Ntest trajectories.
    Xtrain = X[0:Ntrain, 1:401]
    Xval = X[Ntrain:Ntrain+Nval, 1:401]
    Xtest = X[Ntrain+Nval:Ntrain+Nval+Ntest, 1:401]

    Ytrain = Y[0:Ntrain, :]
    Yval = Y[Ntrain:Ntrain+Nval, :]
    Ytest = Y[Ntrain+Nval:Ntrain+Nval+Ntest, :]
    
    #calculating the Fourier coefficients
    XtrainF = np.fft.fft(Xtrain)
    XvalF = np.fft.fft(Xval)
    XtestF = np.fft.fft(Xtest)
    
    #Splitting the coefficients up into their real and imaginary components
    xtrain = np.zeros((XtrainF.shape[0], 2*XtrainF.shape[1]))
    xval = np.zeros((XvalF.shape[0], 2*XvalF.shape[1]))
    xtest = np.zeros((XtestF.shape[0], 2*XtestF.shape[1]))

    for i in range(XtrainF.shape[0]):
        for j in range(XtrainF.shape[1]):
            xtrain[i, 2*j] = XtrainF[i,j].real
            xtrain[i, 2*j + 1] = XtrainF[i,j].imag
        
    for i in range(XtestF.shape[0]):
        for j in range(XtestF.shape[1]):
            xtest[i, 2*j] = XtestF[i,j].real
            xtest[i, 2*j + 1] = XtestF[i,j].imag
            
    for i in range(XvalF.shape[0]):
        for j in range(XvalF.shape[1]):
            xval[i, 2*j] = XvalF[i,j].real
            xval[i, 2*j + 1] = XvalF[i,j].imag
            
    
    return(xtrain, xval, xtest, Ytrain, Yval, Ytest)
```

##### sub-Ohmic

```python
Ysubs = Ysub[:,[4]]

print('Maximum sub-Ohmic s is', np.max(Ysubs))
print('Minimum sub-Ohmic s is', np.min(Ysubs))

Ysubclass = np.zeros((Ysub.shape[0], 5))

for i in range(Ysub.shape[0]):
    if 0 < Ysub[i,4] <= 0.1:
        Ysubclass[i,0] = 1
    if 0.1 < Ysub[i,4] <= 0.2:
        Ysubclass[i,1] = 1
    if 0.2 < Ysub[i,4] <= 0.3:
        Ysubclass[i,2] = 1
    if 0.3 < Ysub[i,4] <= 0.4:
        Ysubclass[i,3] = 1
    if 0.4 < Ysub[i,4] <= 0.5:
        Ysubclass[i,4] = 1
        
xtrain_sub_sclass, xval_sub_sclass, xtest_sub_sclass, Ytrain_sub_sclass, Yval_sub_sclass, Ytest_sub_sclass = fouriertrainvaltest(Xsub, Ysubclass, 2560, 320, 320)
```

```python
#defining the model
model_sub_sclass = tf.keras.Sequential()
model_sub_sclass.add(tf.keras.layers.Dense(250, input_dim = (np.shape(xtrain_sub_sclass)[1]), activation='sigmoid'))
model_sub_sclass.add(tf.keras.layers.Dense(80 , activation = 'sigmoid'))
model_sub_sclass.add(tf.keras.layers.Dense(5, activation = 'softmax'))

#setting the optimiser equal to the Adam optimiser with learning rate = 0.0001
opt = tf.keras.optimizers.Adam(learning_rate = 0.0001)

#compliling the model
model_sub_sclass.compile(optimizer=opt, loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])

#training the model
history_sub_sclass = model_sub_sclass.fit(xtrain_sub_sclass, Ytrain_sub_sclass, epochs = 10000, validation_data = (xval_sub_sclass, Yval_sub_sclass), batch_size = np.shape(xtrain_sub_sclass)[0], verbose=1)
```

Given a dataset containing N trajectories, let $y_{i}$ represent the true value of a parameter corresponding to the $i-th$ trajectory in the dataset and let $\hat{y}_i$ denote the predicted value of the same. Then the coefficient of determination or $R^2$ value is a measure that provides information about the goodness of fit of a model. In the context of regression it is a statistical measure of how well the regression line approximates the actual data. It is therefore important when a statistical model is used either to predict future outcomes or in the testing of hypotheses. The $R^2$ value is defined as

\begin{align}
R^2 & = 1 - \frac{\textrm{sum squared regression(SSR)}}{\textrm{total sum of squares (SST)}} \\
& = 1 - \frac{\sum \left( y_{i} - \hat{y}_i \right)^2 }{\sum \left( y_i - \bar{y}\right)^2 } 
\end{align}

(Applied Linear Statistical Models - Kutner, Nachtsheim, Neter, Li)

```python
def r2(y_true, y_pred):
    ss_res = 0
    ss_tot = 0
    
    y_true_mean = np.sum(y_true)/y_true.shape[0]
    for i in range(y_true.shape[0]):
        ss_res += (y_true[i] - y_pred[i])**2
        ss_tot += (y_true[i] - y_true_mean)**2
        
    return(1 - ss_res/ss_tot)
```

```python
## Custom R-squared metric function

def r_square(y_true, y_pred):
    ss_res = tf.reduce_sum(tf.square(y_true - y_pred))
    ss_tot = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
    return(1 - ss_res/ss_tot)
```
